```python
# Exception handling in file operations

# Types of errors:
# 1. FileNotFoundError: Raised when trying to open a file that does not exist.
# 2. IOError: Raised when an I/O operation fails, such as when trying to read from a file that is not open.
# 3. PermissionError: Raised when trying to access a file without the necessary permissions.
# 4. IsADirectoryError: Raised when trying to open a directory as a file.
# 5. UnicodeDecodeError: Raised when trying to read a file with an incompatible encoding.
# 6. ValueError: Raised when an operation or function receives an argument that has the right type but an inappropriate value, such as trying to write to a file opened in read mode.
# 7. OSError: A general error that can occur during file operations, such as issues with the file system or hardware.
# 8. EOFError: Raised when the end of a file is reached unexpectedly while reading.
# 9. MemoryError: Raised when a file is too large to be read into memory.
# 10. TypeError: Raised when an operation or function is applied to an object of inappropriate type, such as trying to write a non-string object to a file opened in text mode.
# 11. NameError: Raised when trying to access a file that has not been defined or is not in the current scope.
# 12. FileExistsError: Raised when trying to create a file that already exists without using the appropriate mode (e.g., 'x' for exclusive creation').
```

```python
# Exception handling using try-except blocks
try:
    with open('non_existent_file.txt', 'r') as file:
        content = file.read()
except FileNotFoundError:
    print("The file does not exist.")
except IOError:
    print("An I/O error occurred while trying to read the file.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
# Excepton handling with multiple except blocks for nameerror and valuerror example
try:
    x = int("abc")
except NameError:
    print("The variable is not defined.")
except ValueError:
    print("Invalid value.")
```

```python
# Division by zero example
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero.")
```

```python
# TypeError example
try:
    with open('example.txt', 'w') as file:
        file.write(123)  # This will raise a TypeError because write() expects a string
except TypeError:
    print("Cannot write a non-string object to a file opened in text mode.")
```

```python
# other error examples can be added similarly by using appropriate except blocks for each error type.

try:
    with open('example.txt', 'rb') as file:
        content = file.read()
        print(content.decode('utf-8'))  # Attempt to decode the content as UTF-8
except UnicodeDecodeError:
    print("The file contains characters that cannot be decoded with the specified encoding.")
```

```python
# try except else block example
try:
    with open('example.txt', 'r') as file:
        content = file.read()
except FileNotFoundError:
    print("The file does not exist.")
else:
    print("File read successfully. Content:")
    print(content)
```

```python
# try except else finally block example
try:
    with open('example.txt', 'r') as file:
        content = file.read()
except FileNotFoundError:
    print("The file does not exist.")
else:
    print("File read successfully. Content:")
    print(content)
finally:
    print("This block will always execute, regardless of whether an exception occurred or not.")
```
