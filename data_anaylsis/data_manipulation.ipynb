```python
# Data Manipulation with Pandas
import pandas as pd

# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}

df = pd.DataFrame(data)

# Display the DataFrame
print("Original DataFrame:")
print(df)
```

```python
sales_df = pd.read_csv('sales_data_practice.csv')

# Display the first few rows of the sales DataFrame
print("\nSales DataFrame:")
print(sales_df.head())
```

```python
# describbe the sales DataFrame
print("\nSales DataFrame Description:")
print(sales_df.describe())
```

```python
# Datatypes of the sales DataFrame
print("\nSales DataFrame Datatypes:")
print(sales_df.dtypes)
```

```python
## Handling Missing Values
# Check for missing values in the sales DataFrame
print("\nMissing Values in Sales DataFrame:")
print(sales_df.isnull().sum())
```

```python
# head of the sales DataFrame
print("\nHead of Sales DataFrame:")
print(sales_df.head())
```

```python
# fill missing values with 0
sales_df_filled = sales_df.fillna(0)
print("\nSales DataFrame with Missing Values Filled:")
print(sales_df_filled.head())
```

```python
# filling missing values with mean of the column
sales_df['sales_filled_mean'] = sales_df['Price'].fillna(sales_df['Price'].mean())
print("\nSales DataFrame with Missing Values Filled with Mean:")
print(sales_df.head())
```

```python
sales_df.dtypes
```

```python
# Renaming columns
sales_df.rename(columns={'Price': 'Unit_Price'}, inplace=True)
print("\nSales DataFrame with Renamed Columns:")
print(sales_df.head())
```

```python
# change datatype of 'Unit_Price' column to float
sales_df['Unit_Price'] = sales_df['Unit_Price'].astype(float)
print("\nSales DataFrame with 'Unit_Price' Column as Float:")
print(sales_df.dtypes)
```

```python
sales_df['Price'] = sales_df['Unit_Price'] * 3
```

```python
# data Aggregation and Grouping

# Grouping by 'Category' and calculating the mean of 'Price'
category_mean = sales_df.groupby('Category')['Price'].mean()
print("\nMean Price by Category:")
print(category_mean)
```

```python
# Grouping by two categorical columns and calculating the sum of 'Price'
category_product_sum = sales_df.groupby(['Category', 'Product'])['Price'].sum()
print("\nSum of Price by Category and Product:")
print(category_product_sum)
```

```python
# aggregate multiple functions on a column
category_agg = sales_df.groupby('Product')['Price'].agg(['mean', 'sum', 'count'])
print("\nAggregated Price by Product:")
print(category_agg)
```

```python
## Merging and Joining DataFrames
# Create another DataFrame to merge with sales_df
data2 = {
    'Product': ['A', 'B', 'C', 'D', 'E'],
    'Category': ['X', 'Y', 'X', 'Y', 'X'],
    'Stock': [100, 150, 200, 250, 300]
}
data2_df = pd.DataFrame(data2)

# Merge sales_df with data2_df on 'Product' and 'Category'
merged = pd.merge(sales_df, data2_df, on=['Product', 'Category'], how='left')
print("\nMerged DataFrame:")
print(merged.head())
```

```python
pd.merge(sales_df, data2_df, on=['Product', 'Category'], how='inner')
```

```python
pd.merge(sales_df, data2_df, on=['Product', 'Category'], how='outer')
```

```python
pd.merge(sales_df, data2_df, on=['Product', 'Category'], how='left')
```

```python
pd.merge(sales_df, data2_df, on=['Product', 'Category'], how='right')
```
