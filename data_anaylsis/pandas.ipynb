```python
# Pandas
# Pandas is a powerful library in Python for data manipulation and analysis. It provides data structures like Series and DataFrame, which are built on top of NumPy arrays and offer a wide range of functionalities for handling and analyzing data efficiently.
# In this notebook, we will cover the basics of Pandas, including how to create Series and DataFrames, perform basic operations, and handle missing data.

# Difference between Series and DataFrame:
# A Series is a one-dimensional labeled array that can hold any data type (integers,
# floats, strings, etc.). It is similar to a column in a spreadsheet or a database table. Each element in a Series has an associated index, which can be used to access and manipulate the data.
# A DataFrame, on the other hand, is a two-dimensional labeled data structure that can hold multiple Series (columns) of different data types. It is similar to a table in a relational database or a spreadsheet. Each column in a DataFrame is a Series, and the DataFrame itself
# has an index for rows and columns, allowing for easy data manipulation and analysis.
```

```python
import pandas as pd
import numpy as np

# Series

data = [10, 20, 30, 40, 50]
series = pd.Series(data)
print("Series:\n", series)
print("Type of series:", type(series))
```

```python
# Create a Series from a dictionary
data_dict = {'a': 1, 'b': 2, 'c': 3}
series_from_dict = pd.Series(data_dict)
print("\nSeries from dictionary:\n", series_from_dict)
print("Type of series from dictionary:", type(series_from_dict))
```

```python
data = [10, 20, 30, 40, 50]
index = ['a', 'b', 'c', 'd', 'e']
series_with_index = pd.Series(data, index=index)
print("\nSeries with custom index:\n", series_with_index)
print("Type of series with custom index:", type(series_with_index))
```

```python
# dataframe
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Age': [25, 30, 35, 40, 45],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
}
df = pd.DataFrame(data)
print("\nDataFrame:\n", df)
print("Type of DataFrame:", type(df))
```

```python
print("\nDataFrame as NumPy array:\n", np.array(df))
print("Type of DataFrame as NumPy array:", type(np.array(df)))
```

```python
# datframe with lists of dictionaries
data = [
    {'Name': 'Alice', 'Age': 25, 'City': 'New York'},
    {'Name': 'Bob', 'Age': 30, 'City': 'Los Angeles'},
    {'Name': 'Charlie', 'Age': 35, 'City': 'Chicago'},
    {'Name': 'David', 'Age': 40, 'City': 'Houston'},
    {'Name': 'Eve', 'Age': 45, 'City': 'Phoenix'}
]
df_from_list_of_dicts = pd.DataFrame(data)
print("\nDataFrame from list of dictionaries:\n", df_from_list_of_dicts)
print("Type of DataFrame from list of dictionaries:", type(df_from_list_of_dicts))
```

```python
# data manipulation with dataframe

# add a new column 'Salary' to the DataFrame
df['Salary'] = [50000, 60000, 70000, 80000, 90000]
print("\nDataFrame after adding 'Salary' column:\n", df.head(5))
```

```python
# Removing a column from the DataFrame

df.drop('Salary', axis=1, inplace=True)
print("\nDataFrame after removing 'Salary' column:\n", df.head(5))
```

```python
# Add age to the column 'Age' by 1 year for all rows
df['Age'] = df['Age'] + 1
print("\nDataFrame after adding 1 year to 'Age' column:\n", df.head(5))
```

```python
# Drop based on row index
df.drop(1, axis=0, inplace=True)
print("\nDataFrame after dropping row with index 1:\n", df.head(5))
```
