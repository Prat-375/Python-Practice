```python
# Reading Data from differennt sources

import pandas as pd

from io import StringIO # StringIO is used to read data from a string as if it were a file

Data = '{"employee_name": "James", "email": "james@gmail.com", "job_profile": [{"title1":"Team Lead", "title2":"Sr. Developer"}]}'
df = pd.read_json(StringIO(Data))
print(df)
```

```python
df.to_json() # Convert DataFrame to JSON string
```

```python
df.to_json(orient='index') # Convert DataFrame to JSON string with index orientation
```

```python
df.to_json(orient='records') # Convert DataFrame to JSON string with records orientation
```
