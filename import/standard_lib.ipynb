```python
# Python's standard library provides a wide range of modules and functions that can be imported and used in your code.
#  The import statement is used to bring these modules into your current namespace, allowing you to access their functionality.
#  In this notebook, we will explore how to import modules and use their functions effectively.

import array
# The array module provides an efficient way to store and manipulate arrays of basic data types.
#  It is particularly useful when you need to work with large datasets or perform numerical computations.
#  By importing the array module, you can create and manipulate arrays of integers, floats, and other data types, which can be more memory-efficient than using lists.
arr = array.array('i', [1, 2, 3, 4, 5])
print(arr)
```

```python
import random
# The random module provides functions for generating random numbers and performing random operations.
#  By importing the random module, you can generate random integers, floats, and even shuffle sequences. This is particularly useful in scenarios such as simulations, games, and data sampling.
random_number = random.randint(1, 100)
print(random_number)

print(random.choice(['apple', 'banana', 'cherry']))
```

```python
## Files and Directories Access
import os
# The os module provides a way of using operating system dependent functionality like reading or writing to the file system. 
# By importing the os module, you can interact with the file system, manage directories, and perform various operations such as 
# creating, deleting, and renaming files and directories.
current_directory = os.getcwd()
print(current_directory)

os.mkdir('new_directory')
print(os.listdir('.'))
```

```python
## High level file operations
import shutil
# The shutil module provides a higher level interface for file operations, such as copying and moving files and directories. 
# By importing the shutil module, you can easily perform operations like copying files, moving files, and 
# even removing entire directories with their contents.
shutil.copyfile('source.txt', 'destination_file.txt')
```

```python
# Data Serialization
import json
# The json module provides functions for working with JSON data. By importing the json module, 
# you can easily serialize Python objects to JSON format and deserialize JSON data back into Python objects.

data = {'name': 'Alice', 'age': 30, 'city': 'New York'}
json_data = json.dumps(data)
print(json_data)

parsed_data = json.loads(json_data) # Convert JSON string back to Python dictionary
print(parsed_data)

print(type(json_data)) # Output: <class 'str'>
print(type(parsed_data)) # Output: <class 'dict'>
```

```python
## CSV File Handling
import csv
# The csv module provides functionality to read from and write to CSV files. 
# By importing the csv module, you can easily handle CSV data in your Python programs.

with open('data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Name', 'Age', 'City'])
    writer.writerow(['Alice', 30, 'New York'])
    writer.writerow(['Bob', 25, 'Los Angeles'])

with open('data.csv', mode='r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

```python
## Datetine and Time Handling
import datetime
# The datetime module provides classes for manipulating dates and times.
# By importing the datetime module, you can work with dates and times in both simple and complex ways,
#  such as formatting, arithmetic operations, and time zone handling.

current_datetime = datetime.datetime.now()
print(current_datetime)

formatted_date = current_datetime.strftime('%Y-%m-%d %H:%M:%S')
print(formatted_date)

parsed_datetime = datetime.datetime.strptime(formatted_date, '%Y-%m-%d %H:%M:%S')
print(parsed_datetime)

yesterday = current_datetime - datetime.timedelta(days=1)
print(yesterday)
```

```python
## sleep function
import time
# The time module provides various time-related functions. 
# By importing the time module, you can perform operations such as pausing the execution of your program for a
#  specified amount of time, measuring the time taken by code execution, and working with timestamps.

print("Program will sleep for 2 seconds...")
time.sleep(2)
print("Program has resumed after sleeping.")
```

```python
## Regular Expressions
import re
# The re module provides support for regular expressions in Python.
# By importing the re module, you can perform powerful pattern matching and text manipulation operations.

pattern = r'\b\w+\b'  # Matches individual words
text = "Hello, world! This is a regular expression example."
matches = re.findall(pattern, text)
print(matches)
```
