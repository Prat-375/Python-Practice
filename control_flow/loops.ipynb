```python
## For Loop
# For loop is used to iterate over a sequence (like a list, tuple, string) or other iterable objects.
# The syntax of a for loop is:
# for variable in sequence:
#     # code block to be executed for each item in the sequence
# Example 1: Iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
# Output: apple
#         banana
#         cherry
```

```python
# For loop can also be used with the range() function to iterate over a sequence of numbers.
# Example 2: Using range() with a for loop
for i in range(5):  # This will iterate from 0 to 4 (5 is not included)
    print(i)
# Output: 0
#         1
#         2
#         3
#         4 
```

```python
for i in range(1, 6):  # This will iterate from 1 to 5 (6 is not included)
    print(i)
# Output: 1
#         2
#         3
#         4 
#         5
```

```python
for i in range(1, 6, 2):  # This will iterate from 1 to 5 with a step of 2
    print(i)
# Output: 1
#         3
#         5
```

```python
for i in range(10, 0, -3):  # This will iterate from 10 to 1 (0 is not included) with a step of -3
    print(i)
# Output: 10
#         7
#         4
#         1
```

```python
## for loop with strings
# You can also use a for loop to iterate over each character in a string.
for char in "Hello":
    print(char)
# Output: H 
#         e
#         l
#         l
#         o 
```

```python
## While Loop
# A while loop is used to execute a block of code repeatedly as long as a given condition is true.
# The syntax of a while loop is:
# while condition:
#     # code block to be executed as long as the condition is true
# Example 1: Using a while loop to print numbers from 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1  # Increment i by 1
# Output: 1
#         2
#         3
#         4
#         5 
```

```python
#ä Example 2: Using a while loop to iterate over a list
fruits = ["apple", "banana", "cherry"]
index = 0
while index < len(fruits):
    print(fruits[index])
    index += 1  # Increment index by 1
# Output: apple
#         banana
#         cherry
```

```python
## Example 3: Using a while loop to calculate the factorial of a number
number = 5
factorial = 1
while number > 1:
    factorial *= number  # Multiply factorial by the current number
    number -= 1  # Decrement number by 1
print(factorial)
# Output: 120 (which is 5 * 4 * 3 * 2 * 1)
```

```python
## Example 4: Using a while loop to create a simple countdown
countdown = 5
while countdown > 0:
    print(countdown)
    countdown -= 1  # Decrement countdown by 1
print("Liftoff!")
# Output: 5
#         4
#         3
#         2
#         1
#         Liftoff!
```

```python
## Example 5: Using a while loop to keep asking for user input until a valid response is given
user_input = ""
while user_input.lower() != "yes":
    user_input = input("Do you want to continue? (yes/no): ")
print("You chose to continue!")
# Output: (User will be prompted to enter "yes" or "no". The loop will continue until the user enters "yes".)
```

```python
# Example 6: while loop with break statement
# The break statement can be used to exit a while loop prematurely when a certain condition is met
count = 0
while True:  # This creates an infinite loop
    print(count)
    count += 1
    if count >= 5:
        break  # Exit the loop when count reaches 5
# Output: 0
#         1
#         2
#         3
#         4
```

```python
## Example 7: While loop with continue statement
# The continue statement can be used to skip the current iteration of the loop and move to the next iteration
for i in range(1, 10):
    if i % 2 == 0:
        continue  # Skip the rest of the code in the loop for even numbers
    print(i)  # This will only print odd numbers
# Output: 1
#         3
#         5
#         7
#         9
```

```python
## While loop with pass statement
# The pass statement is a placeholder that does nothing. It can be used when a statement is syntactically required but you don't want to execute any code.
for i in range(5):
    if i % 2 == 0:
        print(f"{i} is even.") 
        pass  
    else:
        print(f"{i} is odd.")
# Output: 0 is even.
#         1 is odd.
#         2 is even.
#         3 is odd.
#         4 is even.
```

```python
## Nested loops
# You can also use nested loops, which are loops inside other loops.
for i in range(1, 3):  # Outer loop
    for j in range(1, 3):  # Inner loop
        print(f"i: {i}, j: {j}")
# Output: i: 1, j: 1
#         i: 1, j: 2
#         i: 2, j: 1
#         i: 2, j: 2
```

```python
## Example 1: Using nested loops to print a multiplication table
for i in range(1, 4):  # Outer loop for numbers 1 to 3
    for j in range(1, 4):  # Inner loop for numbers 1 to 3
        print(f"{i} x {j} = {i * j}")
# Output: 1 x 1 = 1
#         1 x 2 = 2
#         1 x 3 = 3
#         2 x 1 = 2
#         2 x 2 = 4
#         2 x 3 = 6
#         3 x 1 = 3
#         3 x 2 = 6
#         3 x 3 = 9
```

```python
## Example 3: Using for and while loops together
# You can also use a for loop inside a while loop or vice versa.
# Example: Using a for loop inside a while loop to print a pattern
count = 1
while count <= 3:
    for i in range(count):
        print("*", end="")  # Print stars on the same line
    print()  # Move to the next line after printing stars
    count += 1
# Output: *
#         **
#         ***
```

```python
## Example 4: Using for and while loop print diamond pattern
size = 5
# Upper part of the diamond
for i in range(size):
    print(" " * (size - i - 1) + "*" * (2 * i + 1))
# Lower part of the diamond
for i in range(size - 2, -1, -1):
    print(" " * (size - i - 1) + "*" * (2 * i + 1))
```

```python
## Example 5: Prime number checker using while loop
number = 29
is_prime = True
if number <= 1:
    is_prime = False
else:
    i = 2
    while i <= number ** 0.5:  # Check for factors up to the square root of the number
        if number % i == 0:
            is_prime = False
            break  # Exit the loop if a factor is found
        i += 1
if is_prime:
    print(f"{number} is a prime number.")
else:    print(f"{number} is not a prime number.")
# Output: 29 is a prime number.
```
