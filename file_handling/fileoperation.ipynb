```python
## File Operations
# File handling is an essential part of programming, allowing you to read from and write to files on your computer.
#  In Python, you can use the built-in `open()` function to work with files. 
# Here's a basic overview of how to perform file operations in Python:

# 1. Opening a File
# You can open a file using the `open()` function, which takes the filename and the
# mode in which you want to open the file (e.g., 'r' for reading, 'w' for writing, 'a' for appending).
file = open('example.txt', 'r')  # Open a file for reading

# 2. Reading from a File
# You can read the contents of a file using methods like `read()`, `readline()`, or `readlines()`.
content = file.read()  # Read the entire content of the file
print(content)

# 3. Writing to a File
# To write to a file, you can open it in 'w' mode (which will overwrite the file) or 'a' mode (which will append to the file).
file = open('example.txt', 'w')  # Open a file for writing
file.write('Hello, World!')  # Write a string to the file

# 4. Closing a File
# It's important to close a file after you're done with it to free up system resources. You can use the `close()` method to do this.
file.close()  # Close the file

# 5. Using `with` Statement
# A better way to handle files is to use the `with` statement, which ensures that
# the file is properly closed after its suite finishes, even if an exception is raised.
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# 6. Read line by line
with open('example.txt', 'r') as file:
    for line in file:
        print(line.strip())  # Print each line without extra whitespace
```

```python
# Writing multiple lines to a file
lines = ['First line\n', 'Second line\n', 'Third line\n']
with open('example.txt', 'w') as file:
    file.writelines(lines)  # Write multiple lines to the file
```

```python
# 6. File Modes
# Here are some common file modes you can use with the `open()` function:
# 'r' - Read (default): Opens a file for reading. The file must exist.
# 'w' - Write: Opens a file for writing. If the file already exists, it will be overwritten.
# 'a' - Append: Opens a file for appending. If the file already exists, new data will be added to the end of the file.
# 'b' - Binary: Opens a file in binary mode (e.g., 'rb' for reading binary files, 'wb' for writing binary files).
# 't' - Text: Opens a file in text mode (default). You can combine this with other modes (e.g., 'rt' for reading text files, 'wt' for writing text files).               

# Appending to a file
with open('example.txt', 'a') as file:
    file.write('This line will be appended to the file.\n')  # Append a new line to the file

with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# Binary mode example
with open('example.bin', 'wb') as file:
    file.write(b'\x00\x01\x02\x03')  # Write binary data to the file

with open('example.bin', 'rb') as file:
    binary_content = file.read()
    print(binary_content)
```

```python
# write from source to destination
source_file = 'example.txt'
destination_file = 'copy_of_example.txt'
with open(source_file, 'r') as src, open(destination_file, 'w') as dst:
    for line in src:
        dst.write(line)  # Write each line from source to destination
```

```python
# Writing and reading a file using `with` statement
with open('example.txt', 'w+') as file:
    file.write('This is a test file.\nIt contains multiple lines of text.')
    file.seek(0)  # Move the cursor back to the beginning of the file
    content = file.read()
    print(content)
```

```python
# create a new directory and write a file in it
import os
os.makedirs('new_directory', exist_ok=True)  # Create a new directory
with open('new_directory/new_file.txt', 'w') as file:
    file.write('This file is inside the new directory.')  # Write to a file in the new directory
```

```python
# Listing files in a directory
files = os.listdir('new_directory')
print(files)  # Print the list of files in the new directory
```

```python
# join paths
file_path = os.path.join('new_directory', 'new_file.txt')
print(file_path)  # Print the full path to the file
```

```python
os.getcwd()  # Get the current working directory
```

```python
path = "example.txt"
if os.path.exists(path):
    print(f"The file '{path}' exists.")
else:
    print(f"The file '{path}' does not exist.")
```

```python
# Checking if a path is a file or a directory
if os.path.isfile(path):
    print(f"'{path}' is a file.")
elif os.path.isdir(path):
    print(f"'{path}' is a directory.")
else:
    print(f"'{path}' is neither a file nor a directory.")
```

```python
# Getting absolute path
absolute_path = os.path.abspath(path)
print(f"The absolute path of '{path}' is: {absolute_path}")
```
