```python
# Dictionaries are a collection of key-value pairs. 
# They are unordered, mutable, and indexed. In Python, dictionaries are defined using curly braces {}.
# Creating a dictionary
my_dict = {
    "name": "Alice",
    "age": 30,
    "city": "New York"
}

# Accessing values in a dictionary
print(my_dict["name"])  # Output: Alice
print(my_dict["age"])   # Output: 30
```

```python
# Creating empty dictionary
empty_dict = {}
print(type(empty_dict))  # Output: <class 'dict'>
```

```python
# Creating  empty dictionary using dict() constructor
another_empty_dict = dict()
print(type(another_empty_dict))  # Output: <class 'dict'>
```

```python
# Adding a new key-value pair to the dictionary
my_dict["profession"] = "Engineer"
print(my_dict)
```

```python
# Error handling: Accessing a non-existent key
try:
    print(my_dict["country"])
except KeyError:
    print("Key 'country' does not exist in the dictionary.")
```

```python
# Adding a new key-value pair to the dictionary with already existing key
my_dict["name"] = "Bob"
print(my_dict)  # Output: {'name': 'Bob', 'age': 30, 'city': 'New York', 'profession': 'Engineer'}
```

```python
# Accesing all keys, values, and items in the dictionary using get() method
print(my_dict.get("name"))  # Output: Bob
print(my_dict.get("country", "Key not found"))  # Output: Key not found

# Accessing all keys, values, and items in the dictionary
print(my_dict.keys())    # Output: dict_keys(['name', 'age', 'city', 'profession'])
print(my_dict.values())  # Output: dict_values(['Bob', 30, 'New York', 'Engineer'])
print(my_dict.items())   # Output: dict_items([('name', 'Bob'), ('age', 30), ('city', 'New York'), ('profession', 'Engineer')])
```

```python
# Modifying values in the dictionary
# Dictionaries are mutable, which means we can change their content after they have been created.
my_dict["age"] = 31
print(my_dict)  # Output: {'name': 'Bob', 'age': 31, 'city': 'New York', 'profession': 'Engineer'}

# Removing a key-value pair from the dictionary
del my_dict["city"]
print(my_dict)  # Output: {'name': 'Bob', 'age': 31, 'profession': 'Engineer'}
```

```python
# Difference between shallow copy and deep copy of a dictionary
# A shallow copy of a dictionary creates a new dictionary object but does not create copies of the objects that the dictionary contains. 
# Instead, it references the original objects. 
# 
# A deep copy, on the other hand, creates copies of the objects that the dictionary contains.

# Shallow copy of a dictionary
shallow_copy = my_dict.copy()
print(shallow_copy)  # Output: {'name': 'Bob', 'age': 31, 'profession': 'Engineer'}

# Deep copy of a dictionary
import copy
deep_copy = copy.deepcopy(my_dict)
print(deep_copy)  # Output: {'name': 'Bob', 'age': 31, 'profession': 'Engineer'}
```

```python
# Iterating through a dictionary with .items() method
for key, value in my_dict.items():
    print(f"{key}: {value}")

# Iterating through a dictionary with .keys() method
for key in my_dict.keys():
    print(f"Key: {key}")

# Iterating through a dictionary with .values() method
for value in my_dict.values():
    print(f"Value: {value}")
```

```python
# Nested Dictionaries
nested_dict = {
    "person1": {
        "name": "Alice",
        "age": 30
    },
    "person2": {
        "name": "Bob",
        "age": 31
    }
}
print(nested_dict)
```

```python
# Access nested dictionary values
print(nested_dict["person1"]["name"])  # Output: Alice
print(nested_dict["person2"]["age"])   # Output: 31
```

```python
# Iterating over nested dictionaries
for person, details in nested_dict.items():
    print(f"{person}:")
    for key, value in details.items():
        print(f"  {key}: {value}")
```

```python
# Dictionary Comprehension
# Dictionary comprehension is a concise way to create dictionaries.
squared_dict = {x: x**2 for x in range(5)}
print(squared_dict)  # Output: {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}
```

```python
# Conditional Dictionary Comprehension
# You can add conditions to dictionary comprehensions.
even_squared_dict = {x: x**2 for x in range(10) if x % 2 == 0}
print(even_squared_dict)  # Output: {0: 0, 2: 4, 4: 16, 6: 36, 8: 64}
```

```python
# Practical Examples
# Use a dictionary to count the frequency of characters in a string
text = "hello world"
char_count = {char: text.count(char) for char in set(text)}
print(char_count)  # Output: {'h': 1, 'e': 1, 'l': 3, 'o': 2, ' ': 1, 'w': 1, 'r': 1, 'd': 1}

# Use a dictionary to count the frequency of words in a string
word_count = {word: text.count(word) for word in set(text.split())}
print(word_count)  # Output: {'hello': 1, 'world': 1}
```

```python
# Merge two dictionaries
dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}
merged_dict = {**dict1, **dict2} # The values from dict2 will overwrite those from dict1 if there are duplicate keys
print(merged_dict)  # Output: {'a': 1, 'b': 3, 'c': 4}
```
