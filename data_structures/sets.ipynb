```python
# Sets
# Sets are unordered collections of unique elements. They are mutable, meaning you can add or remove elements from a set after it has been created. 
# Sets are defined using curly braces {} or the set() constructor. 
# They are commonly used for membership testing, removing duplicates from a sequence, and performing mathematical set operations 
# like union, intersection, and difference.

# Create a set
my_set = {1, 2, 3, 4, 5}
print("My set:", my_set)
print("Type of my_set:", type(my_set))
```

```python
# Create an empty set
empty_set = set()
print("Empty set:", empty_set)
print("Type of empty_set:", type(empty_set))
```

```python
# Set with duplicate elements
set_with_duplicates = {1, 2, 2, 3, 4, 4, 5}
print("Set with duplicates (duplicates will be removed):", set_with_duplicates)
```

```python
# Basic Set Operations
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
# Adding an element to a set
set_a.add(5)
print("Set A after adding an element:", set_a)

# Removing an element from a set
set_a.remove(2)
print("Set A after removing an element:", set_a)

# Union of two sets
union_set = set_a.union(set_b)
print("Union of Set A and Set B:", union_set)

# Intersection of two sets
intersection_set = set_a.intersection(set_b)
print("Intersection of Set A and Set B:", intersection_set)

# Difference of two sets
difference_set = set_a.difference(set_b)
print("Difference of Set A and Set B:", difference_set)
```

```python
# Discard an element from a set (does not raise an error if the element is not present)
set_a.discard(10)  # This will not raise an error
print("Set A after discarding an element that is not present:", set_a)
```

```python
# Removing a random element from a set
print("Set A before popping an element:", set_a)
removed_element = set_a.pop()  # First In First Out (FIFO) order is not guaranteed in sets
print("Removed element from Set A:", removed_element)
print("Set A after popping an element:", set_a)
```

```python
# Clearing all elements from a set
set_a.clear()
print("Set A after clearing all elements:", set_a)
```

```python
## Set Membership Testing
# Check if an element is in a set
print("Is 3 in Set B?", 3 in set_b)
print("Is 10 in Set B?", 10 in set_b)
```

```python
# Mathematical Set Operations
# Union of two sets
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
union_set = set_a.union(set_b)
print("Union of Set A and Set B:", union_set) # Output: {1, 2, 3, 4, 5, 6}

# Intersection of two sets
intersection_set = set_a.intersection(set_b)
print("Intersection of Set A and Set B:", intersection_set) #  Output: {3, 4}

# Difference of two sets
difference_set_a = set_a.difference(set_b)
print("Difference of Set A and Set B:", difference_set_a) # Output: {1, 2}
difference_set_b = set_b.difference(set_a)
print("Difference of Set B and Set A:", difference_set_b) # Output: {5, 6}

# intersection_update() method to update a set with the intersection of itself and another set
print("Set A before intersection_update with Set B:", set_a) # Output: {1, 2, 3, 4}
set_a.intersection_update(set_b)
print("Set A after intersection_update with Set B:", set_a) # Output: {3, 4}
```

```python
# Symmetric Difference of two sets
set_a = {1, 2, 3, 4}
set_b = {3, 4, 5, 6}
symmetric_difference_set = set_a.symmetric_difference(set_b)
print("Symmetric Difference of Set A and Set B:", symmetric_difference_set) # Output: {1, 2, 5, 6}
```

```python
# Sets Methods
my_set = {1, 2, 3, 4, 5}
# add() - Adds an element to the set
my_set.add(6)
print("After adding 6 to my_set:", my_set)
# remove() - Removes an element from the set. Raises KeyError if the element is not present.
my_set.remove(2)
print("After removing 2 from my_set:", my_set)
# discard() - Removes an element from the set if it is present. Does not raise an error if the element is not present.
my_set.discard(10)  # This will not raise an error
print("After discarding 10 from my_set (which is not present):", my_set)

# Issubset() - Checks if a set is a subset of another set
set_c = {1, 2}
print("Is set_c a subset of my_set?", set_c.issubset(my_set))

# Issuperset() - Checks if a set is a superset of another set
print("Is my_set a superset of set_c?", my_set.issuperset(set_c))
```

```python
list_random = [1, 2, 3, 4, 4, 5, 5]
set_list = set(list_random)
print("Original list:", list_random)
print("Set created from the list (duplicates removed):", set_list)
```

```python
# Counting unique words in a sentence
sentence = "hello world hello"
words = sentence.split()
unique_words = set(words)
print("Unique words in the sentence:", unique_words)
```
