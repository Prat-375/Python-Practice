```python
# Tuples 
# A tuple is a collection which is ordered and unchangeable. In Python tuples are written with round brackets.

# Create a Tuple
thistuple = ("apple", "banana", "cherry")
print(thistuple)
```

```python
# Tuples allow duplicate values
thistuple = ("apple", "banana", "cherry", "apple", "cherry")
print(thistuple)
```

```python
# Example of Tuple with different data types
thistuple = ("abc", 34, True, 40, "male")
print(thistuple)
```

```python
# Example of tuple with complex example
thistuple = ("abc", 34, True, 40, "male", ["apple", "banana"], ("apple", "banana"))
print(thistuple)
```

```python
# Try mutating a tuple
thistuple[1] = 100
```

```python
# Create an empty tuple and empty list

empty_list = list()
empty_tuple = tuple()

print(type(empty_list))
print(type(empty_tuple))
```

```python
# Convert list into tuple
mylist = ["apple", "banana", "cherry"]
mytuple = tuple(mylist)
print(mytuple)
```

```python
# Convert tuple into list
mytuple = ("apple", "banana", "cherry")
mylist = list(mytuple)
print(mylist)
```

```python
# Accessing tuple items
thistuple = ("apple", "banana", "cherry")
print(thistuple[1]) # Accessing the second item in the tuple
print(thistuple[-1]) # Accessing the last item in the tuple
```

```python
## Tuple Operations
# Concatenation of tuples
tuple1 = ("a", "b", "c")
tuple2 = (1, 2, 3)
tuple3 = tuple1 + tuple2
print(tuple3)
```

```python
## Multiply tuple3 by 2
tuple4 = tuple3 * 2 # This will repeat the elements of tuple3 twice
print(tuple4)
```

```python
##  Immutable nature of tuples
# Tuples are immutable, which means that once a tuple is created, you cannot change its content. 
# However, if a tuple contains mutable objects like lists, you can modify those mutable objects.
mytuple = ("apple", "banana", ["cherry", "date"])
print(mytuple)
mytuple[2][0] = "grape" # Modifying the first element of the list inside the tuple
print(mytuple)
```

```python
## Tuple Methods
# count() method returns the number of times a specified value appears in the tuple.
thistuple = ("apple", "banana", "cherry", "apple")
x = thistuple.count("apple")
print(x)

# index() method finds the first occurrence of the specified value and returns its index.
thistuple = ("apple", "banana", "cherry")
x = thistuple.index("banana")
print(x)
```

```python
# Packing and Unpacking of tuples
# Packing a tuple
fruits = ("apple", "banana", "cherry") # This will create a tuple named fruits with three items: "apple", "banana", and "cherry"
print(fruits)

# Unpacking a tuple
(green, yellow, red) = fruits # This will assign "apple" to green, "banana" to yellow and "cherry" to red
print(green)
print(yellow)
print(red)
```

```python
# Unpacking a tuple with an asterisk
fruits = ("apple", "banana", "cherry", "orange", "kiwi", "melon", "mango")
(green, yellow, red, *other) = fruits
print(green)
print(yellow)
print(red)
print(other)
```

```python
# Nested Tuples
# A tuple can contain another tuple, this is called a nested tuple.
thistuple = ("apple", "banana", "cherry", ("orange", "kiwi", "melon"))
print(thistuple)
print(thistuple[3]) # Accessing the nested tuple
print(thistuple[3][1]) # Accessing the second item of the nested tuple
```

```python
# Iterating through nested tuples
thistuple = ("apple", "banana", "cherry", ("orange", "kiwi", "melon"))
for x in thistuple:
  if isinstance(x, tuple): # Check if the item is a tuple
    for y in x:            # Iterate through the nested tuple
      print(y)
  else:                    # If the item is not a tuple, print it directly
    print(x) 
```
