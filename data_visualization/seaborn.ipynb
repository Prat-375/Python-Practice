```python
# Seaborn
# Seaborn is a powerful data visualization library built on top of Matplotlib. 
# It provides a high-level interface for creating attractive and informative statistical graphics. 
# Seaborn is particularly useful for visualizing complex datasets and performing exploratory data analysis.

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## Basic Plotting with Seaborn
# Seaborn provides several built-in functions for creating different types of plots. Here are some examples:

tips = sns.load_dataset('tips')

tips
```

```python
# Scatter plot
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()
```

```python
# Lineplot
sns.lineplot(data=tips, x='total_bill', y='tip')
plt.title('Line Plot of Total Bill vs Tip')
plt.show()
```

```python
# Categorical plot
sns.barplot(data=tips, x='day', y='total_bill')
plt.title('Bar Plot of Total Bill by Day')
plt.show()
```

```python
# Box plot
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.show()
```

```python
# Violin plot
sns.violinplot(data=tips, x='day', y='total_bill')
plt.title('Violin Plot of Total Bill by Day')
plt.show()
```

```python
# Histogram
sns.histplot(data=tips, x='total_bill', bins=20, kde=True)
plt.title('Histogram of Total Bill')
plt.show()
```

```python
#  KDE Plot
sns.kdeplot(data=tips, x='total_bill', fill=True)
plt.title('KDE Plot of Total Bill')
plt.show()
```

```python
# Pairplot
sns.pairplot(tips)
plt.suptitle('Pair Plot of Tips Dataset', y=1.02)
plt.show()
```

```python
# Heatmap
corr = tips[['total_bill','tip','size']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Tips Dataset')
plt.show()
```

```python
# Customizing Plots
# Seaborn allows you to customize the appearance of your plots using various parameters and themes.
# You can set the style of your plots using sns.set_style() and choose from options like
# 'darkgrid', 'whitegrid', 'dark', 'white', and 'ticks'. 
# You can also customize colors, labels, and other plot elements to make your visualizations more informative and visually appealing.

sns.set_style('whitegrid')
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', palette='Set2')
plt.title('Scatter Plot of Total Bill vs Tip with Custom Style')
plt.show()
```
