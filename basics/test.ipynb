**Question 12:** Write a Python program to demonstrate comparison operators: equal to, not equal to, greater than, less than.

```python
# Your code here
x = 5
y = 10
if x > y:
    print("x is greater than y")
elif x < y:
    print("x is less than y")
elif x == y:
    print("x is equal to y")
elif x != y:
    print("x is not equal to y")
else:
    print("x and y are incomparable")
```

**Question 13:** Write a Python program to demonstrate logical operators: and, or, not.

```python
# Your code here
x = 5
y = 10
if x > y and x != y:
    print("x is greater than y and not equal to y")
elif x < y or x != y:
    print("x is less than y or not equal to y")
elif not (x > y):
    print("x is not greater than y")
else:
    print("x and y are incomparable")
```

**Question 16:** Write a Python program to find the sum of the first n natural numbers.

```python
# Your code here
x = 5
sum = 0
if x > 0: # Check if x is greater than 0 to ensure we are summing natural numbers
    while x > 0: # Loop continues until x becomes 0
        sum = sum + x  # Add the current value of x to sum
        x = x - 1  # Decrement x by 1 to move towards the base case of the loop
print("Sum of first 5 natural numbers:", sum)
```

**Question 17:** Write a Python program to check if a year is a leap year.

```python
# Your code here
year = 2020
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0): # A year is a leap year if it is divisible by 4 but not by 100, or if it is divisible by 400
    print(year, "is a leap year")
else:
    print(year, "is not a leap year")
```

**Question 18:** Write a Python program to reverse a string.

```python
# Your code here
string = "Panorama"
reverse_string = string[::-1] # This slicing technique reverses the string by taking all characters in reverse order
print("Reversed string:", reverse_string)
```

**Question 19:** Write a Python program to check if a string is a palindrome.

```python
# Your code here
string = "dadad"
reversed_string = string[::-1] # This slicing technique reverses the string by taking all characters in reverse order
if string == reversed_string: # Check if the original string is the same as the reversed string, which indicates it is a palindrome
    print(string, "is a palindrome")
else:
    print(string, "is not a palindrome")
```
