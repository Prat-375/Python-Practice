```python
## Variables 
# A variable is a named location in memory that stores a value. In Python, you can create a variable by simply assigning a value to it using the equals sign (=). For example:
x = 10  # This creates a variable named 'x' and assigns it the value 10
y = "Hello, World!"  # This creates a variable named 'y' and assigns it a string value
is_valid = True  # This creates a variable named 'is_valid' and assigns it a boolean value
print("x: ", x)  # Output: 10
print("y: ", y)  # Output: Hello, World!
print("is_valid: ", is_valid)  # Output: True
```

```python
## Naming Conventions
# In Python, there are certain conventions for naming variables. Variable names should be descriptive and should start with a letter or an underscore. They can contain letters, numbers, and underscores, but cannot start with a number. For example:
first_name = "John"  # Valid variable name
last_name = "Doe"  # Valid variable name
_age = 30  # Valid variable name (starts with an underscore)
2nd_place = "Alice"  # Invalid variable name (starts with a number)
first-name = "Bob"  # Invalid variable name (contains a hyphen)
@name = "Charlie"  # Invalid variable name (starts with a special character)
```

```python
## Understanding Variable Types
# In Python, variables can hold values of different types. The type of a variable is determined at runtime based on the value assigned to it. For example:
age = 25  # age is inferred to be of type int
name = "Alice"  # name is inferred to be of type str
is_student = False  # is_student is inferred to be of type bool
height = 5.9  # height is inferred to be of type float 
print(type(age))  # Output: <class 'int'>
print(type(name))  # Output: <class 'str'>
print(type(is_student))  # Output: <class 'bool'>
print(type(height))  # Output: <class 'float'>
```

```python
## Type Checking and Conversion
# You can check the type of a variable using the built-in type() function. Additionally, you can convert between different types using type conversion functions such as int(), str(), float(), and bool(). For example:
num_str = "123"
num_int = int(num_str)  # Convert string to integer
print(num_int)  # Output: 123
print(type(num_int))  # Output: <class 'int'>
num_float = float(num_str)  # Convert string to float
print(num_float)  # Output: 123.0
print(type(num_float))  # Output: <class 'float'>
is_valid_str = "True"
is_valid_bool = bool(is_valid_str)  # Convert string to boolean
print(is_valid_bool)  # Output: True
print(type(is_valid_bool))  # Output: <class 'bool'>
```

```python
# Type Conversion Errors
# If you try to convert a value that cannot be converted to the desired type, it will raise a ValueError. For example:
invalid_num_str = "abc"
try:
    invalid_num_int = int(invalid_num_str)  # This will raise a ValueError
except ValueError as e:
    print("Error:", e)  # Output: Error: invalid literal for int() with base 10: 'abc'
```

```python
# Dynamic Typing
# Since Python is dynamically typed, you can reassign a variable to a value of a different type without any issues. For example:
x = 10  # x is an integer
print(x)  # Output: 10
x = "Now I'm a string!"  # x is now a string
print(x)  # Output: Now I'm a string!
```

```python
# Input and Output
# You can take input from the user using the input() function and display output using the print function. For example:
name = input("Enter your name: ")  # Takes input from the user
print("Hello, " + name + "!")  # Output: Hello, [name]! 
age = int(input("Enter your age: "))  # Takes input and converts it to an integer
print("You are " + str(age) + " years old.")  # Output: You are [age] years old.

## Type of inputs
# The input() function always returns a string, so if you want to work with numbers, you need to convert the input to the appropriate type. For example:
num1 = int(input("Enter the first number: "))  # Convert input to integer
num2 = float(input("Enter the second number: "))  # Convert input to float
result = num1 + num2  # Perform addition
print("The sum is: " + str(result))  # Output: The sum is: [result]
```
