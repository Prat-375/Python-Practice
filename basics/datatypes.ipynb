```python
## Datatypes
## In Python, there are several built-in datatypes that you can use to store and manipulate data. Some of the most common datatypes include:
# 1. int: Represents integer values (e.g., 1, 42, -5)
# 2. float: Represents floating-point numbers (e.g., 3.14, -0.001)
# 3. str: Represents strings, which are sequences of characters (e.g., "Hello, World!")
# 4. bool: Represents boolean values, which can be either True or False
# 5. list: Represents an ordered collection of items (e.g., [1, 2, 3], ["apple", "banana", "cherry"])
# 6. tuple: Represents an ordered, immutable collection of items (e.g., (1, 2, 3), ("apple", "banana", "cherry"))
# 7. dict: Represents a collection of key-value pairs (e.g., {"name": "Alice", "age": 30})
# 8. set: Represents an unordered collection of unique items (e.g., {1, 2, 3}, {"apple", "banana", "cherry"})
# You can use the built-in type() function to check the datatype of a variable. For example:
x = 10
print(type(x))  # Output: <class 'int'>
y = "Hello, World!"
print(type(y))  # Output: <class 'str'>
is_valid = True
print(type(is_valid))  # Output: <class 'bool'>
fruits = ["apple", "banana", "cherry"]
print(type(fruits))  # Output: <class 'list'>
person = {"name": "Alice", "age": 30}
print(type(person))  # Output: <class 'dict'>
unique_numbers = {1, 2, 3}
print(type(unique_numbers))  # Output: <class 'set'>
tuple_example = (1, 2, 3)
print(type(tuple_example))  # Output: <class 'tuple'>
```

```python
## Example 1 - Type
a = 10
b = 3.14

type(a==b)  # Output: <class 'bool'>, because the expression a==b evaluates to False, which is a boolean value.
```

```python
## Example 2 - Adding two different datatypes
x = 5
y = "10"
try:
    result = x + y  # This will raise a TypeError because you cannot add an integer and a string
except TypeError as e:
    print("Error:", e)  # Output: Error: unsupported operand type(s) for +: 'int' and 'str'
```
