```python
# Basic Syntax Rule in Python
# Case Sensitivity: Python is case-sensitive. For example, 'Variable' and 'variable' are considered different identifiers.
# Indentation: Python uses indentation to define blocks of code. Consistent indentation is crucial for the correct execution of the program.
# Comments: Comments in Python start with the '#' symbol and are used to explain code and make it more readable. They are ignored by the Python interpreter.
# Example of a comment:
# This is a comment in Python

'''
This is a multi-line comment in Python.
It can span multiple lines.
'''
```

```python
name="Prateek"
Name="Prafulla"
print(name)  # Output: Prateek
print(Name)  # Output: Prafulla
# Case sensitivity is demonstrated here, as 'name' and 'Name' are treated as different variables.
```

```python
## Indentation Example
if name == "Prateek":
    print("Hello, Prateek!")  # This line is indented and is part of the if block
else:
    print("Hello, someone else!")  # This line is also indented and is part of the else block

## Comments Example
# This is a single-line comment
print("This line will be executed.")  # This comment is inline with code
```

```python
## Line Continuation Example
# Using a backslash for line continuation
total = 1 + 2 + 3 + \
        4 + 5 + 6
print(total)  # Output: 21
```

```python
## Multiple Statements on a Single Line
# Using a semicolon to separate statements
x = 10; y = 20; z = x + y
print(z)  # Output: 30   
```

```python
## Understanding Semantics in Python
# Semantics refers to the meaning of the code. Even if the syntax is correct, if the code does not make logical sense, it will not produce the desired outcome.
# For example, the following code is syntactically correct but semantically incorrect:
a = 10
b = 0
# This will raise a ZeroDivisionError because dividing by zero is not defined
result = a / b
print(result)  # This line will not be executed due to the error above  
```

```python
## Type Inference in Python
# Python is a dynamically typed language, which means that the type of a variable is inferred at runtime based on the value assigned to it. For example:
x = 5  # x is inferred to be of type int
y = "Hello"  # y is inferred to be of type str
print(type(x))  # Output: <class 'int'>
print(type(y))  # Output: <class 'str'> 
```

```python
## Indentation Error Example
# The following code will raise an IndentationError because the second print statement is not indented properly:
if name == "Prateek":
print("Hello, Prateek!") # This line is not indented properly and will cause an IndentationError"
```

```python
## NameError Example
# The following code will raise a NameError because the variable 'age' is not defined before it is used:
print(age)  # This will raise a NameError because 'age' is not defined
age = 25    
```
